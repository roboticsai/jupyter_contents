```python
# Simple plot with matplotlib
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,2,3,4])
plt.show()
```

```python
# Simplet plot with matplotlib using both axes
x = range(5)
plt.plot(x, [x1 ** 2 for x1 in x])
plt.show()
```

```python
import numpy, matplotlib.pyplot as plt
x = numpy.arange(0,5,0.01)
plt.plot(x, [x1**2 for x1 in x])
plt.show()
```

```python
# Multiline plots
import matplotlib.pyplot as plt
x = range(5)
plt.plot(x, [x1 for x1 in x])
plt.plot(x, [x1*x1 for x1 in x])
plt.plot(x, [x1*x1*x1 for x1 in x])
plt.show()
```

```python
# Adding a grid
import matplotlib.pyplot as plt
x = range(5)
plt.figure()
plt.plot(x, [x1 for x1 in x])
plt.plot(x, [x1*2 for x1 in x])
plt.plot(x, [x1*4 for x1 in x])

plt.grid(True)
plt.show()
```

```python
# Limiting the axes
import matplotlib.pyplot as plt
x = range(5)
plt.plot(x, [x1 for x1 in x])
plt.plot(x, [x1*2 for x1 in x])
plt.plot(x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlim(-1,5)
plt.ylim(-1,10)
plt.show()
```

```python
# Adding labels
import matplotlib.pyplot as plt
x = range(5)
plt.plot(x, [x1 for x1 in x])
plt.plot(x, [x1*2 for x1 in x])
plt.plot(x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlim(-1,5)
plt.ylim(-1,10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Adding the title
import matplotlib.pyplot as plt
x = range(5)
plt.plot(x, [x1 for x1 in x])
plt.plot(x, [x1*2 for x1 in x])
plt.plot(x, [x1*4 for x1 in x])
plt.grid(True)
plt.xlim(-1,5)
plt.ylim(-1,10)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Hello matplot!')
plt.show()
```

```python
# Adding a legend
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(5)
plt.plot(x, x, label='linear')
plt.plot(x,x*x, label='square')
plt.plot(x,x*x*x, label='cube')
plt.grid(True)
plt.xlim(-1,5)
plt.ylim(-1,10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
# Saving plots
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(5)
plt.plot(x, x, label='linear')
plt.plot(x, x*x, label='square')
plt.plot(x, x*x*x, label='cubic')
plt.grid(True)
plt.xlim(-1, 5)
plt.ylim(-1, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('hello this is title!')
plt.show()
plt.savefig('plt.png')
```

```python
# Applying custum colors to plot
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(10)
plt.plot(x, x,'r', label='linear')
plt.plot(x, x*x, 'g', label='square')
plt.plot(x, x*x*x, 'b', label='cubic')
plt.grid(True)
plt.xlim(-1,5)
plt.ylim(-1,10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('hello title!')
plt.show()
```

```python
# Plot line style
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(5,10)
plt.plot(x, '-')
plt.plot(x*x,'--')
plt.plot(x*x*x, ':')
plt.show()
```

```python
# Control marker styling
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1, 3, 0.2)
plt.plot(y, '*',
        y+0.5, 'o',
         y+1, 'D',
         y+2, '^',
         y+3, 's'
        )
plt.show()
```

```python
# Types of plot : Histogram ( Matplotlib )
import matplotlib.pyplot as plt, numpy
y = numpy.random.randn(50,100)
plt.hist(y)
plt.show()
```

```python
# Histogram : Seaborn
import seaborn as sns
iris = sns.load_dataset('iris')
sns.distplot(iris['petal_length'],
            label='Petal Length',
             hist=True,
             kde=False
            )
```

```python
# Bar chart : matplot
import matplotlib.pyplot as plt
plt.bar([1,2,3], [1,4,9])
plt.show()
```

```python
# Bar chart : seaborn
from numpy import mean
import seaborn as sns
iris = sns.load_dataset('iris')
sns.barplot(x = 'species',
           y='sepal_width',
            data=iris,
            estimator=mean,
            palette='spring'
           )
```

```python
# plottingh a dictionary using bar chart in matplotlib
import matplotlib.pyplot as plt, numpy
dictionary = {'A':25, 'B':70, 'C':55, 'D':90}
for i, key in enumerate(dictionary):
    plt.bar(i, dictionary[key])
    plt.xticks(numpy.arange(len(dictionary)), dictionary.keys())
plt.show()
```

```python
# Scatter plot : Matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(1000)
y = np.random.rand(1000)
plt.scatter(x,y)
plt.show()
```

```python
# Heat map : matplotlib
import matplotlib.pyplot as plt
import numpy as np
a = np.random.random((16,16))
plt.imshow(a, cmap='hot',
          interpolation='nearest')
plt.show()
```

```python
# Heat map : seaborn
import seaborn as sns
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights, 
           cmap="Y1GnBu",
           linewidths=2,
           xticklabels=True,
           yticklabels=False)
```
